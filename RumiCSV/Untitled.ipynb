# Najboljši rumi 

V projektni nalogi sem zajela podatke vseh rumov na spletni strani https://www.rumratings.com/brands. Zbrala sem naslednje podatke: • Ime • drzava

```python
import pandas as pd
%matplotlib inline

pd.options.display.max_rows = 20
```

```python
rumi = pd.read_csv('podatki.csv')
rumi
```

```python
#Ker je veliko rumov neocenjenih, mi to uniči analizo, zato se odločim da take vržem iz igre pri določenih analizah
rumi1 = rumi[rumi.Ratings > 0]
rumi1
```

```python
#Kakšnih tipov rumov je največ
rumi.groupby('Type').size().sort_values(ascending=False)
```

```python
rumi.groupby('Type').size().sort_values(ascending=False).plot(kind='pie', figsize=(5, 5), fontsize=12)
```

```python
#KOKLIKO RUMOV PRIHAJA IZ RAZLICNIH DRZAV
rumi.groupby('Country').size().sort_values(ascending=False)
```

```python
#Iz katerih drzav pride največ rumov
rumi.groupby('Country').size().sort_values(ascending=False)[rumi.groupby('Country').size()>80].plot(kind='bar', figsize = (10, 7), fontsize = 13)
```

```python
#kateri stili so najbolj prilubljeni v posameznih državah
rumi.groupby(['Country', 'Type']).size().sort_values(ascending=False)
```

```python
# po oceni razporejeni rumi
rumi[['Name','Country', 'Rating']].sort_values('Rating', ascending=False)
```

```python
#Povprečno ocena ruma glede na državo...paz, lahk je samo edn rum in ta ocenjen z 10
rating=rumi.groupby('Country')[['Rating']].mean().sort_values('Rating', ascending=False)
rating
```

```python
#sumljivo mi je, ker ima prvi v zgornji tabeli oceno 10,0000
rumi[rumi['Country'] == 'British Virgin Islands'][['Name','Type','Ratings','Rating']]
```

```python
# najprej uredi padajoče po številu glasov, pri vsaki pa še padajoče po oceni...NAJBOLJSI RUM, najbolsa ocena in najvec glasov!!!
tabela2 = rumi1[['Name','Rating', 'Ratings']].sort_values(['Ratings'], ascending=[False])
tabela2
```

```python
#MEDSEBOJNO OBNAŠANJE OCENE IN STEVILA GLASOV
tabela2.groupby('Rating').mean().plot()
```

```python
#NEKATERI RUMI NIMAJO PODANE CENE, ZATO JIH PRI NASLEDNJI ANALIZI NE UPOŠTEVAM
rumi_s_ceno= rumi[rumi['Price_USD'] != 'None']
rumi_s_ceno
```

```python
#POVEZAVA MED OCENO IN CENO
tabela2 = rumi_s_ceno.sort_values(['Rating'], ascending=[False])
tabela2
```
